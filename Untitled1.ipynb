```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('Placement_Data_Full_Class.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Check missing values and basic info
print("Missing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
print(df.describe())

print("\nPlacement Status Count:")
print(df['status'].value_counts())
```

```python
# Clean data - fill missing salary values
df['salary'] = df['salary'].fillna(0)

# Add placement column (1 = Placed, 0 = Not Placed)
df['placed'] = df['status'].apply(lambda x: 1 if x == 'Placed' else 0)

print("Data Cleaning Done! ✅")
print("\nPlaced vs Not Placed:")
print(df['status'].value_counts())
print(f"\nPlacement Rate: {df['placed'].mean()*100:.1f}%")
```

```python
# Create Dashboard with 4 charts
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Student Placement Analysis Dashboard', 
             fontsize=16, fontweight='bold')

# Chart 1: Placement Status (Pie Chart)
status_counts = df['status'].value_counts()
axes[0,0].pie(status_counts.values, 
              labels=status_counts.index, 
              autopct='%1.1f%%',
              colors=['#5CB85C','#D9534F'],
              startangle=90)
axes[0,0].set_title('Overall Placement Status')

# Chart 2: Gender vs Placement (Bar Chart)
gender_place = df.groupby(['gender','status']).size().unstack()
gender_place.plot(kind='bar', ax=axes[0,1], 
                  color=['#D9534F','#5CB85C'])
axes[0,1].set_title('Gender vs Placement')
axes[0,1].set_xlabel('Gender')
axes[0,1].set_ylabel('Count')
axes[0,1].tick_params(axis='x', rotation=0)

# Chart 3: Degree vs Placement (Bar Chart)
degree_place = df.groupby(['degree_t','status']).size().unstack()
degree_place.plot(kind='bar', ax=axes[1,0],
                  color=['#D9534F','#5CB85C'])
axes[1,0].set_title('Degree Type vs Placement')
axes[1,0].set_xlabel('Degree')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=15)

# Chart 4: MBA Marks vs Salary (Scatter)
placed_df = df[df['salary'] > 0]
axes[1,1].scatter(placed_df['mba_p'], 
                  placed_df['salary'],
                  color='#5C85D6', alpha=0.6)
axes[1,1].set_title('MBA Marks vs Salary')
axes[1,1].set_xlabel('MBA Percentage')
axes[1,1].set_ylabel('Salary')

plt.tight_layout()
plt.savefig('placement_dashboard.png', 
            dpi=150, bbox_inches='tight')
plt.show()
print("Dashboard saved! ✅")
```

```python
# Print Key Findings
print("=" * 40)
print("   STUDENT PLACEMENT ANALYSIS REPORT")
print("=" * 40)

print(f"\n📊 Total Students: {len(df)}")
print(f"✅ Placed: {len(df[df['status']=='Placed'])}")
print(f"❌ Not Placed: {len(df[df['status']=='Not Placed'])}")
print(f"📈 Placement Rate: {df['placed'].mean()*100:.1f}%")

print("\n--- Salary Statistics ---")
placed = df[df['salary'] > 0]
print(f"💰 Average Salary: {placed['salary'].mean():,.0f}")
print(f"💰 Highest Salary: {placed['salary'].max():,.0f}")
print(f"💰 Lowest Salary:  {placed['salary'].min():,.0f}")

print("\n--- Gender Analysis ---")
print(df.groupby('gender')['placed'].mean()*100)

print("\n--- Work Experience Impact ---")
print(df.groupby('workex')['placed'].mean()*100)

# Save to Excel
df.to_excel('placement_report.xlsx', index=False)
print("\n✅ Report saved as placement_report.xlsx")
print("✅ Dashboard saved as placement_dashboard.png")
print("\n🎉 Project Complete!")
```
